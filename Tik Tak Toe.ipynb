```python
ttt=[".",".",".",
     ".",".",".",
     ".",".","."]

winner = None


def view():
    print(ttt[0]+"/"+ttt[1]+"/"+ttt[2])
    print(ttt[3]+"/"+ttt[4]+"/"+ttt[5])
    print(ttt[6]+"/"+ttt[7]+"/"+ttt[8]);

def move(symbol):
    check = False
    
    while check == False:
        position = int(input("put a number between 0 and 8"))
        
        #Check that the position isn't occupated
        if ttt[position] == ".":
            check = True

        else:
            print("This position is already occupated")
    
    ttt[position] = symbol
    
    view()


    
def play():
    global winner
    view()
    
    for i in range(5):
        
        #player 1(x) make a move
        print("X Make a Move")
        symbol = "X"
        move(symbol)
        win()
        print(winner)
        
        
        if winner != "X" and i < 4:
            for j in range(3):
                
                #player 2(o) make a move
                print("O Make a Move")
                symbol = "O"
                move(symbol)
                win()
                print(winner)
                
                if winner == "O":
                    print ("O Wins")
                
                break
        elif winner == "X":
            print ("X Wins")
            
            break
        
        else:
            print("Draw")
            
            break
            
def win():
    
    global winner
    
    horizontalWin()
    verticalWin()
    diagonalWin()
    
# defining victory conditions 
def horizontalWin():
    global winner
    
    for i in range(0,6,3):
        if ttt[i] == ttt[i+1] == ttt[i+2] !=".":
            winner = ttt[i]
            
def verticalWin():
    global winner
    
    for i in range(2):
        if ttt[i] == ttt[i+3] == ttt[i+6] !=".":
            winner = ttt[i]
        
def diagonalWin():
    global winner
    
    if ttt[0] == ttt[4] == ttt[8] != ".":
        winner = ttt[4]
    elif ttt[2] == ttt[4] == ttt[6] != ".":
        winner = ttt[4]
```

```python
play()
```
